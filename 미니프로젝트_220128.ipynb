```python
def get_search_page_url(date,adult,child):
    return "https://flight.naver.com/flights/domestic/GMP-RSU-{}?adult={}&child={}&infant=0&fareType=YC&selectedFlight=".format(date,adult,child)
```

```python
from datetime import datetime
today=datetime.today().strftime('%Y-%m-%d').replace("-","")
today=int(today)
```

```python
import time
```

```python
from tqdm import tqdm_notebook
```

```python
date=int(20220130)
for page in tqdm_notebook(range(today, date+1)):
    time.sleep(3)
```

```python
date=20220130
adult=1
child=0
prod_data_total=[]

for page in tqdm_notebook(range(today, date+1)):
    #1.검색 페이지 이동
    url=get_search_page_url(page,adult,child)
    driver.get(url)
    #페이지 로딩 시간이 필요하므로 2초 딜레이
    time.sleep(10)
    #로드된 페이지의 html 태그 읽어오기
    html=driver.page_source
    soup=BeautifulSoup(html, 'html.parser')
    
    #현재 페이지의 상품 정보 영역 읽어오기
    prod_items=soup.select("div.List_content div.result")
    
    prod_data=[]
    #1페이지 상품명, 스펙정보, 가격정보
    for i in range(len(prod_items)):
        try:
            airline=soup.select("div.List_content div.result div.airline")[i].text
        except:
            title=""
        try:
            time1=soup.select("div.List_content div.result b.time")[i*2].text
        except:
            spec_list=""
        try:
            price_type=soup.select("div.List_content div.result i.domestic_type__30RSq")[i*2].text
        except:
            price=""
        try:
            price=soup.select("div.List_content div.result i.domestic_num__2roTW")[i*2].text
        except:
            price=""

        prod_data.append([airline, time1, price_type,price])
        
    prod_data_total.append(prod_data)
```
