```python
import re
```

```python
#1) 
text=['abcdefg', 'abcde', 'abc']
for i in text:
    if re.match("[a-z]+",i):
        print("Match")
    else:
        print("skip")
```

```python
#2)
text=['abc123xyz', 'define "123"', 'var g = 123;']
for i in text:
    if re.match('[ \w" ";=]+',i):
        print("Match")
    else:
        print("skip")
```

```python
#3)
text=['cat.', '896.', '?=+.', 'abc1']
for i in text:
    if re.match('[\w?+=]+[.]',i):
        print("Match")
    else:
        print("skip")
```

```python
#4)
text=['can', 'man', 'fan', 'dan', 'ran', 'pan']
for i in text:
    if re.match('[cmf][an]+',i):
        print("Match")
    else:
        print("skip")
```

```python
#5)
text=['hog', 'dog', 'bog']
for i in text:
    if re.match('[hd][o][g]',i):
        print("Match")
    else:
        print("skip")
```

```python
#6)
text=['Ana', 'Bob', 'Cpc', 'aax', 'bby', 'ccz']
for i in text:
    if re.match('[A-Z][a-z]+',i):
        print("Match")
    else:
        print("skip")
```

```python
#7)
text=['wazzzzzup', 'wazzzup', 'wazup']
for i in text:
    if re.match('[w][a][z][z]+[u][p]',i):
        print("Match")
    else:
        print("skip")
```

```python
#8)
text=['aaaabcc', 'aabbbbc', 'aacc', 'a']
for i in text:
    if re.match('[a][a][a-z]+',i):
        print("Match")
    else:
        print("skip")
```

```python
#9)
text=['1 file found?', '2 files found?', '24 files found?', 'No files found.']
for i in text:
    if re.match('[0-9]+[ a-z?]+',i):
        print("Match")
    else:
        print("skip")
```

```python
#10)
text=['1.   abc', '2.   abc', '3.           abc', '4.abc']
for i in text:
    if re.match('[0-9][.][ ]+[a-z]+', i):
        print("Match")
    else:
        print("skip")
```

```python
url= "https://news.naver.com/main/read.naver?mode=LSD&mid=shm&sid1=103&oid=052&aid=0001697611"
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
res=urllib.request.urlopen(url)
soup=BeautifulSoup(res, 'html.parser')
```

```python
soup
```

```python
temp = re.compile("[영][상|하][ .0-9]+[도]")
temp.findall(str(soup))
```

```python
s = re.compile("[0-9]+cm")
s.findall(str(soup))
```

```python
tel = re.compile("[\d]{2}[-][\d]{3}[-][\d]{4}")
tel.findall(str(soup))
```

```python
mail = re.compile("\w+@[\w.]+")
mail.findall(str(soup))
```
